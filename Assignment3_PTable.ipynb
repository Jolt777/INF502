```python
#Execute a command (a string) in a subshell

import os
```

```python
#This function provides a title card and a menu of options for the user.

def print_main_window():
    os.system('clear')
    print("\t********************************************")
    print("\t***  The Periodic Table of the Elements  ***")
    print("\t********************************************")
    print("\n\n\n") #3 line brakes
    print("Choose an option as follows: ")
    print("[1] See all information on an element")
    print("[2] View a property of an element")
    print("[3] Add a new element")
    print("[4] Change the attributes of an element")
    print("[5] EXIT")
```

```python
#This function is created to see all the information stored about this particular element, by entering that element's symbol

def information_gatherer(s):
    for sub, pt_list in ptable.items():
        for element in pt_list:
            if element['symbol'] == s:
                print(element)
            elif element['symbol'] != s:
                continue
            else:
                print("Invalid Symbol. Please try again.\n")
    print("\n\n\n")
```

```python
#This function is created to gather a particular property for each element in the PT

def property_gatherer(p):
    for sub, pt_list in ptable.items():
        for element in pt_list:
            print(element['name'] + ": " + str(element[p]))
    print("\n\n\n")
```

```python
#This function allows the user to add a new element

def new_element():
    r1 = input("Enter the name of the element: ")
    r2 = input("In a few words, describe the apprearance of the element: ")
    r3 = input("Enter the atomic mass of the element: ")
    r4 = input("Enter the boiling point of the element: ")
    r5 = input("Name the category of the Periodic Table on which the element belongs to: ")
    r6 = input("Enter the density of the element: ")
    r7 = input("Who or Whom discovered the element: ")
    r8 = input("Enter the melting point of the element: ")
    r9 = input("Enter the molar heat capacity (aka specific heat) of the element: ")
    r10 = input("Who or Whom give the element its name: ")
    r11 = input("What is the number of the element: ")
    r12 = input("What period of the periodic table is the element located: ")
    r13 = input("What group of the periodic table is the element located: ")
    r14 = input("What is the phase of the element at room temperature (solid, liquid, gas, or plasma): ")
    r15 = input("Give a link to the source of this information you found about the element: ")
    r16 = input("Give a link to the spectral imaging of the element (if any): ")
    r17 = input("Give a brief summary about the element: ")
    r18 = input("What is the symbol of the element: ")
    r19 = input("Enter the numeric value of the x-position of the element: ")
    r20 = input("Enter the numeric value of the y-position of the element: ")
    r21 = input("Enter the classic version of the element's electron configuration: ")
    r22 = input("Enter the electron affinity value of the element: ")
    r23 = input("Enter the Pauling eletronegativity value of the element: ")
    new_list = []
    new_result = {"name": r1,
                  "appearance": r2, 
                  "atomic_mass": r3,
                  "boil": r4,
                  "category": r5,
                  "density": r6,
                  "discovered_by": r7,
                  "melt": r8,
                  "molar_heat": r9,
                  "named_by": r10,
                  "number": r11,
                  "period": r12,
                  "group": r13,
                  "phase": r14,
                  "source": r15,
                  "spectral_img": r16,
                  "summary": r17,
                  "symbol": r18,
                  "xpos": r19,
                  "ypos": r20,
                  "electron_configuration": r21,
                  "electron_affinity": r22,
                  "electronegativity_pauling": r23}
    mydict_copy = new_result.copy() #copy() makes a copy of the newly-created dictionary without references
    new_list.append(mydict_copy)
    pt_updated = ptable['elements'] + new_list
    print("You new element has been added to the periodic table\n")
    print(new_result)
    print("\n\n\n")
    print(ptable)
    print("\n\n\n")
```

```python
#This function changes the attribute(s) of an element, occurs when user enters element's symbol

def changes(n):
    for sub, pt_list in ptable.items():
        for element in pt_list:
            if element['symbol'] == n:
                attribute = str(input("Please select an attribute of an element: "))
                rp = input("Please replace the value or make the changes: ")
                element[attribute] = rp
                print("The change has been successfully made!\n")
            elif element['symbol'] != n:
                continue
            else:
                print("Invalid Symbol. Please try again.\n")
    print("\n\n\n")
```
