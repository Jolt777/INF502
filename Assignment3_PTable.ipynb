```python
#This function is created to gather a particular property for each element in the PT

def property_gatherer(p):
    for sub, pt_list in ptable.items():
        for element in pt_list:
            print(element['name'] + ": " + str(element[p]))
    print("\n\n\n")
```

```python
#This function changes the attribute(s) of an element, occurs when user enters element's symbol

def changes(n):
    for sub, pt_list in ptable.items():
        for element in pt_list:
            if element['symbol'] == n:
                attribute = str(input("Please select an attribute of an element: "))
                rp = input("Please replace the value or make the changes: ")
                element[attribute] = rp
                print("The change has been successfully made!\n")
            elif element['symbol'] != n:
                continue
            else:
                print("Invalid Symbol. Please try again.\n")
    print("\n\n\n")
```
